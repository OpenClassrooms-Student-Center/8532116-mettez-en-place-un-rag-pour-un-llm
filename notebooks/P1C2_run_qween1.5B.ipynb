```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# 1. Charger le modèle et le tokenizer
model_name = "jpacifico/Chocolatine-3B-Instruct-DPO-v1.2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# 2. Préparer la question pour le modèle
question = "Quel est le département numéro 29 de la France ?"
inputs = tokenizer(question, return_tensors="pt", padding=True, truncation=True)

# 3. Générer une réponse avec le modèle
output = model.generate(
    inputs["input_ids"],
    attention_mask=inputs["attention_mask"],
    max_new_tokens=50,  # Remplace max_length
    num_return_sequences=1
)

```

```python
# 4. Décoder et afficher la réponse
response = tokenizer.decode(output[0], skip_special_tokens=True)
print("Question :", question)
print("Réponse :", response)
```
